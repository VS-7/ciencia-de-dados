```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('game_dataset.csv')
df
```

# Preenchimento de Elementos Faltantes
Identifique e trate os valores ausentes nas colunas "Classificação de Usuários" e "Idade Recomendada". Você pode optar por preencher esses valores usando métodos como a média, mediana, ou qualquer outra técnica que considerar adequada.

```python
imputer = SimpleImputer(strategy='mean')
df['Classificação de Usuários'] = imputer.fit_transform(df[['Classificação de Usuários']])

imputer = SimpleImputer(strategy='median') #Valor inteiro
df['Idade Recomendada'] = imputer.fit_transform(df[['Idade Recomendada']])
df
```

# Transformação de Variáveis Categóricas
Converta as colunas "Gênero", "Plataforma" e "Lançamento no Brasil" para um formato numérico, utilizando técnicas como one-hot encoding ou label encoding.

```python
df = pd.get_dummies(df, columns=['Gênero', 'Plataforma', 'Lançamento no Brasil'], prefix=['Gênero', 'Plataforma', 'Lançamento no Brasil'])
df
```

# Normalização de Dados
Aplique a normalização nas colunas "Vendas Globais" e "Preço" para que essas variáveis estejam na mesma escala. Você pode utilizar métodos como Min-Max Scaling ou Z-Score.

```python
scaler = MinMaxScaler()
df[['Vendas Globais','Preço']] = scaler.fit_transform(df[['Vendas Globais','Preço']])
df
```

```python
df.to_csv('tableReady.csv', index=False)
```
